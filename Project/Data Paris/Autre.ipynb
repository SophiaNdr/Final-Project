```python
greve['Date']=greve['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
```

```python
greve
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

fig, ax = plt.subplots(figsize=(20, 10))

# Add x-axis and y-axis
ax.scatter(greve['Date'], greve['Journées perdues'], color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Number of People",
       title="Daily frequentation")

plt.show()
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(20, 10))

# Add x-axis and y-axis
ax.bar(greve['Date'],
        greve['Journées perdues'],
        color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Number of days",
       title="Greve over years")

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```

```python
sns.lineplot(greve['Date'], greve['Journées perdues'])
plt.show()
```
